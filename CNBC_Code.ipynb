```python
def yesterday(str):
    data_prev = datetime.strptime(str,'%Y-%m-%d').date()-timedelta(1)
    return(data_prev.strftime('%e %b %Y'))
df2 = pd.read_csv('Stock.csv',encoding='ISO-8859-1', engine='c')
df2['Date_prev']=df2.Date.apply(yesterday)
df2
```

```python
def yesterday(str):
    data_prev = datetime.strptime(str,'%Y-%m-%d').date()-timedelta(1)
    return(data_prev.strftime('%e-%b-%y'))
df3 = pd.read_csv('Stock.csv',encoding='ISO-8859-1', engine='c')
df3['Date_prev']=df3.Date.apply(yesterday)

df3
```

```python
temp = [0]+list(df2['Close'])[0:525]
temp
#df2.shape
df2['pre_close'] = temp
df2=df2.assign(Rise = (df2.Close-df2.pre_close > 0))
df2
```

```python
temp = [0]+list(df3['Close'])[0:525]
temp
#df2.shape
df3['pre_close'] = temp
df3=df3.assign(Rise = (df3.Close-df3.pre_close > 0))
df3
```

```python
df_fn=pd.concat([df2,df3],axis=0)
df_fn
```

```python
data_final_combined=pd.merge(df,df_fn,left_on="date",right_on="Date_prev")
```

```python
data_final_combined.to_csv('CNBC_data_final_combined.csv',encoding='ISO-8859-1')
```
